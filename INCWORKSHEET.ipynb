```python
#warming up exercises
#TASK 1

time_data = [
    (3.5, 2.0, 7.0), (5.0, 1.5, 6.5), (2.5, 3.0, 8.0),
    (4.0, 2.0, 6.0), (1.5, 4.5, 9.0), (3.0, 2.5, 7.5),
    (5.5, 1.0, 6.0), (2.0, 3.5, 8.5), (4.5, 2.0, 7.0),
    (3.0, 3.0, 7.5), (6.0, 1.5, 6.0), (2.5, 4.0, 8.0),
    (4.0, 2.5, 7.0), (5.0, 2.0, 6.5), (3.5, 2.5, 7.0)
]

#1. Create empty lists
low = []
moderate = []
high = []

#2. Classify study hours
for study, entertainment, sleep in time_data:
    if study < 3:
        low.append(study)
    elif 3 <= study <= 5:
        moderate.append(study)
    else:
        high.append(study)

#3. Print lists
print("Low study hours (<3):", low)
print("Moderate study hours (3–5):", moderate)
print("High study hours (>5):", high)
```

```python
#TASK 4

#1
study_hours = []
entertainment_hours = []
sleep_hours = []

#2
for record in time_data:
    study_hours.append(record[0])
    entertainment_hours.append(record[1])
    sleep_hours.append(record[2])

#3 (Calculation)
average_study = sum(study_hours) / len(study_hours)
average_entertainment = sum(entertainment_hours) / len(entertainment_hours)
average_sleep = sum(sleep_hours) / len(sleep_hours)
#a
print("Average hours spent studying:", average_study)
#b
print("Average hours spent on entertainment:", average_entertainment)
#c
print("Average hours spent sleeping:", average_sleep)
```

```python
#RECURSION
#TASK 1

def sum_nested_list(nested_list):
    total = 0
    for element in nested_list:
        if type(element) == list:  #checking element
            total += sum_nested_list(element)
        else:
            total += element
    return total

#Test with a sampke nested list
nested_list = [1, [2, [3, 4], 5], 6, [7, 8]]
result = sum_nested_list(nested_list)
print("Total sum:", result)
```

```python
#TASK 2

def generate_permutations(s):
    if len(s) <= 1:
        return [s]

    permutations = []
    for i, char in enumerate(s):
        remaining = s[:i] + s[i+1:]
        for perm in generate_permutations(remaining):
            permutations.append(char + perm)

    return list(set(permutations))  #Remove duplicates

#1
print(generate_permutations("abc"))
#2
print(generate_permutations("aab"))
```

```python
#TASK 3

#1
def calculate_directory_size(directory):
    total_size = 0
    for key, value in directory.items():
        if type(value) == dict:
            total_size += calculate_directory_size(value)
        else:
            total_size += value
    return total_size

#Sample directory structure
directory_structure = {
    "file1.txt": 200,
    "file2.txt": 300,
    "subdir1": {
        "file3.txt": 400,
        "file4.txt": 100
    },
    "subdir2": {
        "subsubdir1": {
            "file5.txt": 250
        },
        "file6.txt": 150
    }
}

#2
print("Total directory size:", calculate_directory_size(directory_structure))
```
